## Problem 1

Write a program that will simulate the Wright-Fisher model for a diploid population segregating two alleles, A and a.

Then, run 1000 simulations starting at $p_0 = .5$ to estimate the distribution of $p_t$ at times $t = \frac{N}{2}$ and $t=N$ for $N$ ranging from $10^2$ to $10^5$. What do your results suggest about the behavior of the process when time is measured in units of $N$ generations and $N \rightarrow \infty$?

### Answer:

It appears that the process will converge to one allele being fixed when the time grows large. However, it appears that more than $2N$ generations are needed to observe full fixation in a range of simulations, as only around 7% of the simulations achieved fixation in each population size.

### Methods:

First, the program. I will need to import a few libraries like `NumPy` and `Pickle`. `NumPy` is just a high-speed numerical library for python, and `Pickle` allows me to store data in a fast, simple binary format to avoid memory limits while processing markov chains.

```python
import numpy as np
import pickle as pk
from collections import defaultdict
import matplotlib.pyplot as plt
from depickle import *
import os
import pandas as pd
%matplotlib inline
```

With these, first I program the simulator for the Wright-Fisher model. I used the binomial sampling technique over $2N$ individuals mentioned in the practical notes. This code does one simulation for a population size `pop`, with alleles `anames` that have initial probabilities `initprobs`, for `gens` generations. This code is the basic unit of computation, which was meted out to a pool and done in parallel. 

```python
def wf(pop, anames, initprobs, gens):
    results = defaultdict(dict)
    i = 0
    results[i] = {key:val for key,val in zip(anames, initprobs)}
    p = results[i][anames[0]]
    
    while gens > i:
        p = np.random.binomial(2*pop, p)/ float(2*pop)
        i += 1
        results[i] = {key:val for key,val in zip(anames, [p, 1-p])} 
        
    return results
```

But, storing $S$ copties of this simulation in memory all at once may prove to be too costly. To avoid this, I use the pickle library to save the results of each simulation to a binary file, further allowing easy simulations.

## Problem 3

Use the results from Problem 2 to estimate the mean time to fixation as a function of $N$ and $p_0$. 

### Answer:

```python
n100 = pd.Series([.37, .53, .62, .69 ,.733, .70, .64, .53, .37])
n1000 = pd.Series([.353, .502, .633, .722, .749, .698, .593, .499, .341])
n10000 = pd.Series([.2531, .5228, .6148, .6736, .7307, .6849, .5920, .4858, .3392])
p = pd.Series([.1, .2, .3, .4, .5, .6, .7, .8, .9])
```

```python
df = pd.DataFrame([n100, n1000, n10000, p]).transpose()
```

My observations of fixation time standardized by population size are shown in the following table.

```python
df.columns = ['n100','n1000','n10000','p']
df
```

This shows a function that increases in mean time to fixation up to $.5$, then decreases as a function of $p$ as it moves further away from $.5$. A naive model with relatively good fit would be a parabola centered at $.5$ and censored once it reaches $0$. Symbolically, we can represent this as a relationship between $\tau$, stopping time, $N$, population size, and $p_0$, the starting probability. A simple function that fits the data very well is:

$$ \frac{\tau}{N} = -2.4\left(p_0 - .5\right)^2 + .72$$

And, this function is positive, nonzero in the range $p_0 \in [0,1]$, which makes sense. But, the probability of fixation should approach zero as $p_0$ approaches 1 and 0, and this form has positive nonzero fixation time at $p_0 = 1$ or $0$. However, this appears good enough for the values of $p$ tested, in the range $p_0 \in [.1, .9]$. 

A plot of this function, dashed in black, is shown below. 

```python
sp = np.linspace(0, 1, num=10000)
a = plt.figure(figsize=(5*1.6,5))
plt.plot(df.p, df.n100)
plt.plot(df.p, df.n1000)
plt.plot(df.p, df.n10000)
plt.plot(sp, [-2.4*(x - .5)**2 + .72 for x in sp], 'k--')
plt.title('Data and Fitted Parabola')
plt.xlabel('Starting Probability, $p_0$')
plt.ylabel('Time to Fixation: $\\frac{G_t}{N}$')
```

## Problem 4

Modify the program to model a population with two sexes in which there are $N_f$ females and $N_m$ males in each generation. Assume that each individual in $\mathcal{G}_{t+1}$ inherits one chromosome from a randomly chosen female alive and the other chromasome from a randomly chosen male. 

Then, run 1000 simulations lasting a single generation to estimate the mean and variance of the change in allele frequency, $p_1 - p_0$ for initial $p$ values $.1, .2, ..., .9$, given that the initial frequency is the same in males and females. Do this for $(N_f, N_m) = (950, 50), (900, 100), (800, 200),$ and $(500, 500)$ and use your results to discuss how skewed sex ratios affect genetic drift. 

### Answer:

The more skewed the sex ratio, the larger the swings possible in the generation. This means that the frequencies of the sub-populations likely float independently in this model, and small populations are already vulnerable to large shocks to the genetic process between generations. 

It seems like sampling within the skewed sex ratios means that a few lucky samples from the smaller group have disproportionate impact on the subsequent samples. Given the data, there was almost a difference in magnitude between males and females when the sex ratios are rather skewed, but converge to relative similarity when the sex ratios get close to even. From this, we could claim that skewed sex ratios increase volatility in the allele probabilities of the smaller group. That is, for this case, the variance and mean allele probability was very different, with no directional effect but higher volaitlity, for simulations with sex ratios skewed towards women than in simulations where sex ratios that were even.

### Methods:

The exact algorithm used to simulate the sex-based simulations as almost identical to the form used above. However, independent probabilities for males and females were kept. Since the simulation instructions did not mention how to vary the population counts, and the alleles being referred to were not stated to be sexually determining alleles, I kept the percentages of women and men constant throughout the simulation. This allows the subpopulations of males and females to have different allele frequency rates while having constant subpopulation sizes.

The algorithm is shown below:

```python
def sexsim_ps(fpct, pop, gens):
    mpct = round(1 - fpct, 2)
    plist = np.arange(.1,1,.1)
    for p0 in plist:
        p_f = p_m = p0
        mpct = 1 - fpct
        i = 0
        pkl = open('./sexsims/' + str(int(fpct*100)) + 'p' + str(int(100*p0)) + '.pkl', 'ab')
        results = []
        while gens > i:
            i +=1
            p_f = round(np.random.binomial(int(2 * fpct * pop), p_f) / (2 * fpct * pop), 3)
            p_m = round(np.random.binomial(int(2 * mpct * pop), p_m) / (2 * mpct * pop), 3)
            results.append((p_f, p_m))
        pk.dump(results, pkl)
        pkl.close()
```

Exxentially, after setting up the constants controlling for female population size, male population size, and the initial probabilities, we simulate over some arbitrary number of simulations, sampling each binomial distribution for males and females independently. This is what "builds in" the independence discussed in the answer. It would be interesting to see what the simulation results would look like if the alleles in question were also sex-determining. Just from a computational perspective, how to set up that simulation would be more interesting, at least, as this process was very simple to make parallel.
